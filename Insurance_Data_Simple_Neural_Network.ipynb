```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("/content/sample_data/insurance_data.csv")
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[['age']],df['bought_insurance'],test_size=0.2,random_state=2)
```

```python
x_train_scaled = x_train/100
x_test_scaled = x_test/100
```

**SIMPLE NUARAL NETWORK (NO HIDDEN LAYER)**

```python
model = keras.Sequential([
      keras.layers.Dense(1, input_shape=(1,), activation='sigmoid' , kernel_initializer='ones',bias_initializer='zeros')
])

model.compile(
    optimizer= 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)

model.fit(x_train_scaled, y_train, epochs=5000)
```

```python
model = keras.Sequential([
        keras.layers.Dense(5,input_shape=(1,), activation='relu'),
        keras.layers.Dense(1,activation='sigmoid')
])

model.compile(
    optimizer= 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)

model.fit(x_train_scaled, y_train, epochs=1000)
```
